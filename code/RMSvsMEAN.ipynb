```python
import numpy as np
import matplotlib.pyplot as plt

#Function
period = 8 * np.pi
x = np.linspace(0,period, 10000)
y = np.sin(x)
#arithmetic mean
arithmetic_mean_array = np.full_like(x, sum(y) / len(y))
#rms
rms_array = np.full_like(x, ((np.sum(np.square(y)) / len(y)) ** .5))

plt.plot(x,arithmetic_mean_array)
plt.plot(x,rms_array)
plt.plot(x,y)
plt.legend(['Arithmetic Mean','Root Mean Square','F = sin(x)'])
plt.xlabel('x')
plt.ylabel('Value')
plt.show()

print(f"RMS: {rms_array[1]:.3f}")
print(f"Arithmetic Mean: {abs(arithmetic_mean_array[1]):.3f}")
```
