```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simpson

maxX = 20

x = np.linspace(0,maxX,20)
p = 0.30617536265 * np.exp(-((x - 2)**2) / 4)
horiz_arr = np.linspace(-1,1,40)

#data
for x_val, y_val in zip(x,p):
    plt.quiver(x, np.zeros_like(x), np.zeros_like(x), p, angles='xy', scale_units='xy', scale=1, color='purple')

#x axis and plotted first moment

line2, = plt.plot((np.full_like(horiz_arr, 2.22)), horiz_arr, color='red')
plt.plot(x,np.zeros_like(x),color='black')

plt.legend([line2],["First Moment"])
plt.ylim(-.1,max(p)*1.1)
plt.xlabel("x")
plt.ylabel("P(x)")
plt.show()
```
