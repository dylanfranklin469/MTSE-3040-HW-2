```python
import numpy as np
import matplotlib.pyplot as plt

T = np.linspace(1,3000,10000)
m = 2.325 * 10 **-26
k = 1.38 * (10 ** -23)
mp = (2 * k * T / m) ** .5
ave = (8 * k * T / (np.pi * m)) ** .5
ratioFn = ave / mp
marker = round(ratioFn[0], 5)



plt.plot(T, mp)
plt.plot(T,ave)
plt.plot(T, ratioFn)
plt.legend(['Most Probable','Average Value','Ratio'])
plt.xlabel("Temperature (K)")
plt.ylabel("Value")
plt.title("Most Probable vs Average Value Trend Over Varying Temperature")
plt.show()

print("The ratio between the two is", marker)
for element in ratioFn:
    rouElement = round(element,5)
    if rouElement != marker:
        print(marker)
        print("Differing Value!!")
```
